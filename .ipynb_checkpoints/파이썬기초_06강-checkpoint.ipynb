```python
i=1 #처음 i의 초기값 1로 설정
while i < 5: #if문처럼 조건은 괄호없이 :로 마무리 
    i+=1  #i에 먼저 1을 더하므로 5까지 출력된다
    print(i) # i값을 출력
```

```python
i=0 #반복을 카운트 할 변수에 0을 대입하여 초기 설정.
sum=0 #누적할 변수 세팅해주기
while i <100 : #0부터 100까지 실행하게 한다
    i+=1 #i는 1씩 누적 증가.
    sum = sum + i #sum에 증가하는 i를 더하며 누적 증가.
print("sum=",sum) #while문이 끝나고 결과 sum을 출력한다.
```

```python
i=0
sum=0
while i <100 : #0부터 99까지 실행하게 한다
    sum = sum + i #sum에 증가하는 i를 더하며 누적 증가.
    i+=1 #i는 1씩 누적 증가. i가 sum을 계산한 후 증가하므로 마지막 i가 100일때 반복이 끝나므로 4950
print("sum=",sum)
```

```python
i=1 #변수 i의 초기값 1로 설정
while True: #무한반복 실행
    j=i #변수 j 초기값 1로 설정
    while True: #무한반복 실행
        if(j==10): #j가 10이 되면 현 반복문 탈출
            break
        print( "*",end="") 
        j+=1 #반복문을 도는 j에 누적 더하기
    if(i==10): 
        break #i가 10이 되면 탈출
    print(end="\n") #줄 바꿔주고
    i+=1 #i에 1누적 더하기
```

```python
Mon=1 #시작하는 달 변수 Mon을 1월부터 설정하기
for LastDay in [31,28,31,30,31,30,31,31,30,31,30,31]: #각 순서별로 달의 마지막 날을 int값으로
    #컬렉션에 저장
    print(Mon,"월 :",end="") #1씩 증가하는 월을 출력, end잡아줘서 줄바꿈 없이
    for Day in range(1,LastDay+1): #컬렉션의 요소를 받는 변수 Day를 1부터 시작해서 컬렉션의 순번
        #12까지(11+1) 범위를 잡아주어 한달 한달 마지막 날을 출력하게
        print(Day,end=",") #엔드값을 콤마로 잡아주기
    print(end="\n") #한 달 출력 다하면 
    Mon+=1 #반복 끝나면 다음 달로 넘어가게 누적 더하기
```

```python
for i in range(2,10,3): #구구단 단 수 출력. 2부터 시작해서 증가는 3씩. 9까지. 첫줄은 2 5 8
    for j in range(2,10): # 구구단 둘째 자리 2~9까지 반복하게 만듦
        print(i,"*",j,"=",i*j,end="\t") # 구구단 포맷 출력. 엔드값은 줄바뀌지 않게 탭으로 구분
        print(i+1,"*",j,"=",i*j,end="\t") #옆 단으로 넘어가면서 i+1 해주기
        if(i+2!=10): #8*2줄에서 10단은 없으므로 
            print(i+2,"*",j,"=",i*j,end="\n") #4단 7단이면 줄바꿔서 이어서 출력
        else: 
            print(end="\n") #10단이 된다면 9단 찍고 바로 줄바꿈
    print(end="\n") #단 사이의 한줄 공백
```

```python
for i in range(2,5): #반복문 돌 변수 i는 2부터 4까지. i i+3 i+6단 순이다
    for j in range(2,10): #오른쪽에 곱해지는 값 2~9까지 반복문 설정
        print(i,"*",j,"=",i*j,end="\t") #첫단 i단
        print(i+3,"*",j,"=",i*j,end="\t")#둘째단 i+3단
        if(i+6!=10): 
            print(i+6,"*",j,"=",i*j,end="\n") #셋째단 i+6단. 이때 10단은 줄바꿈해 노출력
        else:
            print(end="\n") #조건이 8,9이면 줄바꿈해서 이어서 출력
    print(end="\n") #단 사이의 1줄 공백
```
